```python
# üìò Notebook: 1_VideoSetupAndPreprocessing.ipynb
# Purpose: Load input video, display a sample frame, setup paths

# üì¶ Step 1: Import required libraries
import cv2
import matplotlib.pyplot as plt
import os
import warnings
from termcolor import colored  # For console colored output

# üîá Suppress warnings for cleaner output
warnings.filterwarnings('ignore')

print(colored("üì¶ Step 1: Libraries loaded successfully", "green", attrs=['reverse']))
```

```python
# üìÅ Step 2: Define input video path
video_path = r"D:\RoadRadar-Real-Time-Speed-Detection-System\input\sample_video.mp4"
```

```python
# ‚úÖ Step 3: Check if the video file exists
if not os.path.exists(video_path):
    raise FileNotFoundError(f"Video not found at: {video_path}")
else:
    print("‚úÖ Input video found:", video_path)
```

```python
# üé¨ Step 4: Load and read the first frame of the video
cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()
cap.release()

if not ret:
    raise Exception("‚ùå Failed to read frame from video")
```

```python
# üîÅ Step 5: Convert the BGR frame to RGB for matplotlib display
frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

```

```python
# üñºÔ∏è Step 6: Display the first frame from the video
plt.figure(figsize=(12, 6))
plt.imshow(frame_rgb)
plt.title("üì∏ Sample Frame from Trimmed Video")
plt.axis("off")
plt.show()
```
